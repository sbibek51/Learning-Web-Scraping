## Extract raw html

```python
response=requests.get(template, headers= headers)
```

```python
soup = BeautifulSoup(response.text,'html.parser')
```

```python
print(soup)
```

```python
# Extract information
title = soup.find('h3', class_='title-name').get_text(strip=True)
author = soup.find('p', class_='title-author').get_text(strip=True).replace('by ', '')
format_ = soup.find('span', class_='title-format-badge').get_text(strip=True)
link = soup.find('a', href=True)['href']

print("Title:", title)
print("Author:", author)
print("Format:", format_)
print("Link:", link)
```
