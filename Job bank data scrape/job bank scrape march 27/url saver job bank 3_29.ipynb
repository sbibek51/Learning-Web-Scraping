```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import csv
import time
import datetime
from bs4 import BeautifulSoup

# Initialize the Chrome webdriver
driver = webdriver.Chrome()

# Open the base URL
base_url = 'https://www.jobbank.gc.ca/jobsearch/jobsearch?sort=M&fprov=MB&fprov=NB&fprov=NL&fprov=NS&fprov=NT&fprov=NU&fprov=PE&fprov=SK&fprov=YT' 
driver.get(base_url)

# Initialize counters
url_counter = 0
button_counter = 0

# Define a function to scrape job links
def scrape_job_links():
    # Get the page source after Selenium has rendered the page
    page_source = driver.page_source

    # Use BeautifulSoup to parse the HTML
    soup = BeautifulSoup(page_source, 'html.parser')

    # Find all job links on the page
    job_links = soup.find_all("a", class_="resultJobItem")

    # Create a set to store the extracted links to avoid duplicates
    links_set = set()

    # Iterate through the job links and extract the URLs
    for link in job_links:
        job_url = link.get("href")
        # Append the base URL if the job URL is relative
        if not job_url.startswith("http"):
            job_url = "https://www.jobbank.gc.ca" + job_url
        links_set.add(job_url)
    return links_set

# Define a function to click the "Show more results" button
def click_show_more():
    global button_counter
    try:
        show_more_button = driver.find_element(By.XPATH, "//button[@id='moreresultbutton']")
        show_more_button.click()
        button_counter += 1  # Increment the button counter
        print(f"\rShow more button clicked: [{button_counter}]", end='', flush=True)  # Print in same line without newline
        time.sleep(3)  # Wait for the page to load
        return True
    except:
        print("\nNo more Show more button")
        return False

# Scrape initial job links
all_job_links = scrape_job_links()
url_counter += len(all_job_links)

# Keep clicking "Show more results" and scraping links until the button is not available
while click_show_more():
    links = scrape_job_links()
    all_job_links.update(links)
    url_counter += len(links)

# Define the path for the CSV file
num_links_saved = len(all_job_links)
current_date = datetime.date.today().strftime("%m_%d")
csv_file_path = f"job_links_small_provinces_{num_links_saved}_{current_date}.csv"

# Write all extracted links to a CSV file
with open(csv_file_path, "w", newline="", encoding="utf-8") as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(["Job Link"])
    for link in all_job_links:
        writer.writerow([link])

# Print completion message
print("\nJob links have been saved to", csv_file_path)

# Close the webdriver
driver.quit()
```
