```python

# Convert the 'posted_date' column to datetime objects
df['Posting Date'] = pd.to_datetime(df['Posting Date'])

# Calculate the difference between each date and today's date
df['posted_date_mm_dd_yyyy'] = df['Posting Date'].dt.strftime('%m-%d-%Y')
df['days_since_posted'] = (datetime.now() - df['Posting Date']).dt.days

# Print the DataFrame with the new column
```

```python
df.head()
```

```python
# df.describe(percentiles=[0.01,.1,.2,.3,.4,.5,.6,.7,.8,.9,0.99])
```

```python
df['month'] = df['Posting Date'].dt.month

# Plotting  histogram
plt.figure(figsize=(10, 6))
df['month'].plot(kind='hist', bins=12, color='skyblue', edgecolor='black', linewidth=1.2)


plt.title('Histogram of Counts by Month')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

```python
# Count the occurrences of each unique value in the 'Building Type' column
building_type_counts = df['Building Type'].value_counts()

# Plot histogram
plt.figure(figsize=(10, 6))
building_type_counts.plot(kind='bar', color='skyblue', edgecolor='black', linewidth=1.2)

# Customize the plot
plt.title('Histogram of Building Types')
plt.xlabel('Building Type')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

```python
df['Price']
```

```python
df['Price_numeric'] = df['Price'].str.extract(r'(\$[\d,]+)')
df['Price Note'] = df['Price'].str.extract(r'(\$[\d,]+)(.*)')[1]

```

```python
df.head()
```

```python
df['Price Note']
```

```python
# Find rows where 'Utilities Included' column is NaN or empty and 'Price Note' column is not empty
result = df[(df['Utilities Included'].isna() | df['Utilities Included'].eq('')) & df['Price Note'].ne('')]

print(result)
```

```python
# Count rows where 'Utilities Included' column is NaN or empty and 'Price Note' column is not empty
filtered = df[(df['Utilities Included'].isna() | df['Utilities Included'].eq('')) & df['Price Note'].notna() & df['Price Note'].ne('')]

```

```python
filtered
```

```python
unique_price_notes = filtered['Price Note'].value_counts()

print("Unique Price Notes:")
print(unique_price_notes)
```

```python
# now we can replace the useful informartion from price note to the Utilities included column
```

```python
# Impute values from 'Price Note' column to 'Utilities Included' column where 'Utilities Included' is NaN or empty
df.loc[(df['Utilities Included'].isna() | df['Utilities Included'].eq('')) & (df['Price Note'].notna() & df['Price Note'] != ''), 'Utilities Included'] = df['Price Note']
```
